Probablility Mass Function: divide through the total count to normalize the distribution

```python
import thinkstats2
views = thinkstats2.Hist(df.views, label='views')
thinkplot.Hist(views)
thinkplot.Config(xlabel='views', ylabel='Count')
```

```python
n = views.Total()
pmf = views.Copy()
for x, freq in views.Items():
    pmf[x] = freq / n
```

The result is a Probability Mass Function (PMF).

```python
import thinkplot
thinkplot.Hist(pmf)
thinkplot.Config(xlabel='Views', ylabel='PMF')
```

Correlation

```python
import thinkstats2

def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr
```

```python
Corr(views, likes)
```
